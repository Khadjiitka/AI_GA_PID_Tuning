```python
import random

# Уявимо, що ми протестували 3 різних варіанти PID і отримали для них оцінки:
tested_population = [
    {"kp": 1.2, "ki": 0.05, "kd": 0.1, "fitness": 45.2},  # Погано тримався
    {"kp": 2.1, "ki": 0.10, "kd": 0.4, "fitness": 80.14}, # Наш реальний лідер від Юрія!
    {"kp": 3.5, "ki": 0.25, "kd": 0.8, "fitness": 12.3}   # Жахлива тряска
]

print(" Результати першого покоління перед еволюцією:")
for i, ind in enumerate(tested_population, 1):
    print(f"Кандидат №{i}: PID({ind['kp']}, {ind['ki']}, {ind['kd']}) -> Оцінка: {ind['fitness']}")

# 1. СЕЛЕКЦІЯ (Selection): сортуємо кандидатів за оцінкою і вибираємо найкращого
tested_population.sort(key=lambda x: x['fitness'], reverse=True)
parent = tested_population[0] # Наш найкращий кандидат (80.14 балів)

print(f"\n ШІ вибрав найкращого 'батька' для нового покоління: PID({parent['kp']}, {parent['ki']}, {parent['kd']})")

# 2. МУТАЦІЯ (Mutation): ШІ бере найкращі коефіцієнти і трохи змінює їх, щоб знайти ідеал
new_population = []
for i in range(3): # Генеруємо 3 нових покращених кандидатів
    # Додаємо випадкову невеличку мутацію (від -0.2 до +0.2) до кожного коефіцієнта батька
    new_kp = round(parent['kp'] + random.uniform(-0.2, 0.2), 3)
    new_ki = round(parent['ki'] + random.uniform(-0.02, 0.02), 3)
    new_kd = round(parent['kd'] + random.uniform(-0.05, 0.05), 3)
    
    # Захист від від'ємних значень (коефіцієнти не можуть бути менше 0)
    new_population.append({
        "kp": max(0.0, new_kp),
        "ki": max(0.0, new_ki),
        "kd": max(0.0, new_kd)
    })

print("\n ШІ успішно згенерував ПОКОЛІННЯ №2 (Покращені кандидати на основі лідера):")
for i, child in enumerate(new_population, 1):
    print(f" Новий Кандидат №{i}: Kp={child['kp']}, Ki={child['ki']}, Kd={child['kd']}")
```

```python
import random

class GeneticPIDOptimizer:
    def __init__(self, population_size=5, mutation_rate=0.2):
        self.population_size = population_size
        self.mutation_rate = mutation_rate

    def init_population(self):
        """Генерує першу випадкову популяцію коефіцієнтів PID."""
        population = []
        for _ in range(self.population_size):
            candidate = {
                "kp": round(random.uniform(0.5, 3.5), 3),
                "ki": round(random.uniform(0.0, 0.5), 3),
                "kd": round(random.uniform(0.0, 1.0), 3)
            }
            population.append(candidate)
        return population

    def mutate(self, parent):
        """Створює мутованого нащадка на основі найкращого батька."""
        new_kp = round(parent['kp'] + random.uniform(-self.mutation_rate, self.mutation_rate), 3)
        new_ki = round(parent['ki'] + random.uniform(-self.mutation_rate * 0.1, self.mutation_rate * 0.1), 3)
        new_kd = round(parent['kd'] + random.uniform(-self.mutation_rate * 0.2, self.mutation_rate * 0.2), 3)
        
        return {
            "kp": max(0.0, new_kp),
            "ki": max(0.0, new_ki),
            "kd": max(0.0, new_kd)
        }

    def evolve(self, tested_population):
        """
        Приймає список кандидатів з їхніми оцінками (fitness),
        вибирає найкращого і створює нове покоління.
        """
        # Сортуємо від найкращого (найбільший бал) до найгіршого
        tested_population.sort(key=lambda x: x['fitness'], reverse=True)
        best_parent = tested_population[0]
        
        new_generation = []
        # Залишаємо самого переможця без змін (елітизм), щоб не втратити ідеал
        new_generation.append({
            "kp": best_parent["kp"],
            "ki": best_parent["ki"],
            "kd": best_parent["kd"]
        })
        
        # Заповнюємо решту місць у популяції мутованими нащадками лідера
        while len(new_generation) < self.population_size:
            mutated_child = self.mutate(best_parent)
            new_generation.append(mutated_child)
            
        return new_generation
```
